```python
import csv
from sklearn.tree import DecisionTreeRegressor
import numpy as np
all_data = []
with open("Seasons_Stats.csv", "r") as f:
    reader = csv.reader(f)
    first = 1
    for row in reader:
        all_data.append(row)
statistics = all_data[0]
print( statistics)
```

```python
def data_from_years():
    data = []
    for person in all_data[1:]:
        if person[1] != "":
            data.append(person)
    return data

all_players = data_from_years()
count = 0;
for row in all_players[-10:]:
    count = count + 1
    print("Count: %s, %s" % (count, row))
```

```python
def people_of_years():
    data_year = all_players
    all_people = []
    games = {}
    game_count = 0
    for player in data_year:
        if player[2] not in games:
            games[player[2]] = int(player[statistics.index("G")])
        else:
            games[player[2]] += int(player[statistics.index("G")])
    person_dict = {}
    for i in range(len(data_year)-1):
        player = data_year[i]
        person_dict = {}
        for statistic in statistics[3:]:
            if "blan" not in statistic and player[statistics.index(statistic)] != "":
                    person_dict[statistic] = player[statistics.index(statistic)] 
        all_people.append((player[2],person_dict))
    return all_people  

players_year = people_of_years()
count = 0;
for row in all_players[-10:]:
    count = count + 1
    print("Count: %s, %s" % (count, row))
```

```python
import random
def model_mvp(stats_to_use):
    xlist = []
    for player1 in players_year:
        moze = 1
        player = player1[1]
        player1_list = []
        if 30 < int(player["G"]) and int(player["G"]) is not "":
            for i in range(len(stats_to_use)):
                if stats_to_use[i] in player:
                    player1_list.append(float(player[stats_to_use[i]]))
                else:
                    moze = 0;
            if(moze == 1):
                xlist.append(player1_list)
    random.shuffle(xlist)
    return xlist     
```

```python
adv = set(["VORP","OWS","WS","WS/48","FT","FTA","PTS","FG","DWS"])

adv_stats = model_mvp(list(adv))
```

```python
def how_good_new(which_stats_fit,which_stats_predict):
    model = DecisionTreeRegressor(max_depth = 485)
    x_stats = [x[:-1] for x in which_stats_fit]
    y_stats = [y[-1] for y in which_stats_fit]
    model.fit(np.array(x_stats), y_stats)
    x1_stats = [x[1:-1] for x in which_stats_predict]
    answer = []
    for j in range(len(x1_stats)):
        prediction = model.predict(np.asarray(x1_stats[j]).reshape(1, -1))
#         if prediction[0] > .0025:
#             print which_stats[:int(.8*length_data)][j][-1], prediction[0]
        answer.append([float(prediction.tolist()[0]),new_stats[j][0]])
    answer.sort()
    return answer

order = how_good_new(adv_stats,new_stats)
for thing in order:
    print ("Igrac:%s, koef:%s" % (thing[0],thing[1]))
        
```
